```python
#import libraries
import numpy as np 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras import regularizers
```

```python
#use iris data
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelBinarizer
```

```python

iris = load_iris()
x = iris.data
y=iris.target
print(x)
print(y)
```

```python
#use labelbinarizer
data1 =LabelBinarizer()
y = data1.fit_transform(y)
```

```python
#Normalize features
scalar=StandardScaler()
Scaled=scalar.fit_transform(x)
```

```python
print(Scaled)
```

```python
X_train,X_test,y_train,y_test = train_test_split(Scaled,y,test_size=0.2,random_state=42)
```

```python
y_test.shape
```

```python
X_train.shape
```

```python
import numpy as np 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras import regularizers
```

```python
def model_creation():
    model = Sequential([
        Dense(10,input_shape=(4,),activation="relu"),
        Dense(3,activation="softmax")
    ])
    model.compile(optimizer=SGD(learning_rate=0.01),loss=CategoricalCrossentropy(),metrics=["accuracy"])
    return model
```

```python
#model without backpropogation
model1 = model_creation()
pred1 = model1.predict(X_test)
pred1
```

```python
#model with backpropogation
model2 = model_creation()
model2.fit(X_train,y_train,epochs= 100)
pred2 = model2.predict(X_test)
pred2
```

```python
loss,accuracy = model2.evaluate(X_test,y_test)
print(loss)
print(accuracy)
```

```python
loss,accuracy = model1.evaluate(X_test,y_test)
print(loss)
print(accuracy)
```
